```python
import pandas as pd
```

```python
df_path = "/Users/zwanran/Desktop/STAT27420/Final/STAT27420-final-code/Dataset/Crime.dta"
df = pd.read_stata(df_path)
df = df[(df['cohort'] >1957) & (df['cohort']<1963)]
```

# Variance in Covariates
Covariates = origin and district (24 districs in total)

We only have data for the covariates during 1958-1962. Also, we posit that such controls, as demographic distribution, are less likely (than potential controls like socioeconomic status) to vary drastically across years.

```python
df.iloc[:,range(23, 51)].apply(["mean","median", "min", "max", "var"])
```

```python
for yr in range(1958,1962):
     print([df[df['cohort']==yr].iloc[:,range(23, 51)].apply(["mean","median", "min", "max", "var"])])
```

# Variance in Outcome

```python
varlist = ["crimerate", "income", "unemployment", "crimrecordi", "sexual", "murder", "threat", "drug", "whitecollar", "arms"]
df[varlist].apply(["mean","median", "min", "max", "var"])
```

```python
for yr in range(1958,1962):
     print(df[df['cohort']==yr][varlist].apply(["mean","median", "min", "max", "var"]))
```
